
###Analyze Best Selling Amazon Books
> with Pandas



```python
# About the .csv file (provided by Kaggle) :
# The .csv file contains 550 books. The seven columns include:
# Name: Book name
# Author: Book author
# User Rating: Amazon user rating (0.0 - 5.0)
# Reviews: Number of user reviews
# Price: Book price (as of 2020)
# Year: The year(s) it ranked
# Genre: Fiction or non-fiction
```

```python
import pandas as pd

#using dataframe, load the spreadsheet
df = pd.read_csv('/content/bestsellers.csv')
```

```python
df
```

```python
#Explore the Data - to get an overview
print(df.head())
print(df.shape)
print(df.describe())
```

```python
#Cleaning the Data
df.drop_duplicates(inplace=True)
```

```python
#Renaming the Columns (using Dictionaries):
df.rename(columns=
 {"Name": "Title",
 "Year": "Publication Year",
 "User Rating": "Rating"}, inplace=True)
```

```python
df.dtypes
```

```python
#Converting Data Types (as needed):
df['Price'] = df['Price'].astype(float)
```

```python
#Run Analysis
#1. Find out the authors with most books
author_counts = df['Author'].value_counts()
print(author_counts)
```

```python
#2. Average rating by Genre (using GroupBy)
avg_rating =  df.groupby('Genre')['Rating'].mean()
print(avg_rating)
```

```python
#Export top selling authors to a .csv file
author_counts.head(10).to_csv("Top_Authors.csv")

# Export average rating by genre to a CSV file
avg_rating.to_csv("Avg_Rating_by_Genre.csv")
```
